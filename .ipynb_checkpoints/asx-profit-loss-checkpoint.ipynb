```python
# Plot each ticker as seperate line graphs

# 4DS
# Define data
fourds_time = fourds_df['Date']
fourds_open = fourds_df['Open']
fourds_high = fourds_df['High']
fourds_low = fourds_df['Low']
fourds_close = fourds_df['Close']
fourds_adj_close = fourds_df['Adj Close']
fourds_volume = fourds_df['Volume']

# Font families
font_1 = {'family':'helvetica','color':'black','size':16}
font_2 = {'family':'helvetica','color':'black','size':12}

# Plotting
plt.figure(figsize=(15,5))
plt.plot(fourds_time, fourds_adj_close, color='blue', label="Closing price")

# Rolling 10-day average / STD
def get_rolling_mean(values, window):
    return values.rolling(center=False, window=10).mean()
def get_rolling_std(values, window):
    return values.rolling(center=False, window=10).std()

# Bollinger bands
def get_bollinger_bands(rollingMean, rollingSTDEV):
    # Calculate upper and lower bollinger bands
    upper_band = rollingMean + rollingSTDEV * 2
    lower_band = rollingMean - rollingSTDEV * 2
    return upper_band, lower_band

def plot_data(fourds_df, title="Stock Prices"):
    # Add labels
    ax = fourds_df.plot(title=title, fontdict = font_1)
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    #plt.show()
    
    # Calculate daily returns
    daily_returns = fourds_adj_close
    plot_data(daily_returns, title = "Daily Returns")

    # Generate Bollinger Bands window=10 means 10-day moving avarage
    # Using get_rolling_mean to retrieve standard moving average, not exponential
    rolling_mean_MSFT = get_rolling_mean(fourds_df['Adj Close'], window=10)

     # Generate rolling standard deviation (10-day)
    rolling_STDEV_MSFT = get_rolling_std(four_df['Adj Close'], window=10)

    # Generate upper and lower bands
    upper_band, lower_band = get_bollinger_bands(rolling_mean_MSFT, rolling_STDEV_MSFT)

    # plot MSFT values, rolling mean and Bollinger Bands
    ax = fourds_df['Adj Close'].plot(title="Bollinger Bands", label='MSFT')
    rolling_mean_MSFT.plot(label='Rolling mean', ax=ax)
    upper_band.plot(label='upper band', ax=ax)
    lower_band.plot(label='lower band', ax=ax)

    # Add labels 
    ax.legend(loc='bottom left')
    ax.set_xlabel("Date")
    ax.set_ylabel("Price")
    plt.show()


#plt.title("4DS price performance, 1 May 2020 - 23 February 2022", fontdict = font_1)
#plt.xlabel("Date", fontdict = font_2)
#plt.ylabel("Price", fontdict = font_2)
#plt.legend()

# Aesthetics

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()

#smooth = fourds_df['Adj Close'].rolling(window=10, label="Rolling 10-day average").mean()
#smooth.plot()

# Save and show
# plt.savefig("/Users/minoperic/Data Analytics/Data Investigations/vanda-and-young-finalists-2021/Images/publishers.png", dpi=300)
#plt.show()
```
