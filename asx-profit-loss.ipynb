```python
# Import dependencies and setup
import matplotlib.pyplot as plt
import matplotlib as matplotlib
import pandas as pd
import datetime as dt
import matplotlib.dates as mdates
```

```python
# Read .csv for each stock and store as DataFrame

# 4DS
fourds_file_path = "Resources/4DS.csv"
fourds_df = pd.read_csv(fourds_file_path)
fourds_df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

# Define and add 'TP', 'STD', 'MA-TP', 'BOLU' and 'BOLD' columns
fourds_df['TP'] = (fourds_df['Close'] + fourds_df['Low'] + fourds_df['High'])/3
fourds_df['std'] = fourds_df['TP'].rolling(20).std(ddof=0)
fourds_df['MA-TP'] = fourds_df['TP'].rolling(20).mean()
fourds_df['BOLU'] = fourds_df['MA-TP'] + 2*fourds_df['std']
fourds_df['BOLD'] = fourds_df['MA-TP'] - 2*fourds_df['std']
fourds_df = fourds_df.dropna()
fourds_df
```

```python
# Plot each ticker as seperate line graphs

# 4DS

# Font families
font_1 = {'family':'helvetica','color':'black','size':16}
font_2 = {'family':'helvetica','color':'black','size':12}

# Plotting

plt.figure(figsize=(15,5))
plt.plot(fourds_df["Date"], fourds_df["Adj Close"], color='blue', linewidth=2, label="Closing price")
plt.plot(fourds_df["Date"], fourds_df["MA-TP"], color='red', linewidth=2, label="Rolling 20-day average")
plt.plot(fourds_df["Date"], fourds_df["BOLU"], color='green', linewidth=1, label="BOLU")
plt.plot(fourds_df["Date"], fourds_df["BOLD"], color='orange', linewidth=1, label="BOLD")

d = fourds_df['Date'].values
plt.fill_between(d, fourds_df["BOLU"], fourds_df["BOLD"],
                where=fourds_df["BOLU"] >= fourds_df["BOLD"],
                facecolor='green', alpha=0.1, interpolate=True)

plt.title("4DS price performance, 25 May 2020 — 21 February 2022", fontdict = font_1)
plt.xlabel("Date", fontdict = font_2)
plt.ylabel("Price (AUD)", fontdict = font_2)
plt.legend()

plt.xticks((fourds_df['Date'])[::20], rotation=35)

# Aesthetics

matplotlib.pyplot.grid(visible=True, which='both', axis='both', 
                       color='blue', alpha=0.1, dash_capstyle='butt', dash_joinstyle='bevel', 
                       linestyle='--', linewidth=1)

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()

# Save and show
plt.savefig("/Users/minoperic/Data Analytics/Data Investigations/asx-profit-loss/Images/4DS.png", dpi=300)
plt.show()
```
